# Gold winning agent 

```python
class Environment:
    def __init__(self):
        self.x_limit = 4
        self.y_limit = 6
        self.V = (1,3)
        self.Pit = (3,4)
        self.dest = (3,0)
        self.property = {}
        for x in range(self.x_limit):
            for y in range(self.y_limit):
                if x == 1 and y == 5:
                    self.property[(x,y)] = "G"
                elif (x == 0 and y == 5) or (x == 2 and y == 5):
                    self.property[(x,y)] = "g"
                elif (x == 1 and y == 4):
                    self.property[(x,y)] = "sg"
                elif (x == 0 and y == 3) or (x == 2 and y == 3) or \
                     (x==1 and y ==2) or (x == 1 and y == 3):
                    self.property[(x,y)] = "s"
                elif (x == 3 and y == 4):
                    self.property[(x,y)] = "p"
                elif (x == 3 and y == 0):
                    self.property[(x,y)] = "D"
                elif (x == 3 and y == 3) or (x==3 and y==5) or \
                       (x == 2 and y == 4):
                    self.property[(x,y)] = "b"
                else:
                    self.property[(x,y)] = "e"
    

    def get_state(self, pos):
        adj_cells = {}
        try:
            adj_cells["r"] = self.property[(pos[0]+1,pos[1])]
        except:
            adj_cells["r"] = None
        try:
            adj_cells["l"] = self.property[(pos[0]-1,pos[1])]
        except:
            adj_cells["l"] = None
        try:
            adj_cells["f"] = self.property[(pos[0],pos[1]+1)]
        except:
            adj_cells["f"] = None
        try:
            adj_cells["b"] = self.property[(pos[0],pos[1]+1)]
        except:
            adj_cells["b"] = None
        return adj_cells
        
    def display(self):
        print(self.property)
    
    def check_gold_collected(self,agent):
        if self.property[agent.pos] == "G" :
            print("Agent has collectedd Gold!!")
            agent.gold_collected = True
    
    def check_win(self,agent):
        if agent.pos == self.dest and agent.gold_collected == True:
            print("You are a REAL hero !!!!!")
            return True
        if agent.pos == self.V or agent.pos == self.Pit:
            print("You Die !!")
            return False
        return None
```

```python
import numpy as np
```

```python
class Agent:
    def __init__(self):
        self.pos = (0,0)
        self.gold_collected = False
        
    def move(self,direction="f"):
        if direction == "f":
            self.pos = (self.pos[0], min(self.pos[1]+1,5))
        elif direction == "b":
            self.pos = (self.pos[0], max(self.pos[1]-1,0))
        elif direction == "r":
            self.pos = (min(self.pos[0]+1,3), self.pos[1])
        elif direction == "l":
            self.pos = (max(self.pos[0]-1,0), self.pos[1])
    
    def next_move(self,env):
        # Get dictionary of state of adjecent cells
        #"r" right cell, "l" left cell, "f" forward cell, "b" backward cell
        adj_cell_state = env.get_state(self.pos)
        
        ## Write your logic here to select perfect step
        perfect_step = np.random.choice(["r", "l", "f", "b"])
        
        for prize in ['g','G','D','sg','s']:          # Priorities of cost of each step g>G>D>sg>s
            for move in ["r", "l", "f", "b"]: 
                if (adj_cell_state[move] == prize):
                    perfect_step= move
            
        
        # Make a move
        # Just use self.move() with arg "r", "l", "f", "b"
        self.move(perfect_step)
        
        env.check_gold_collected(self)
        result = env.check_win(self)
        return result
        
        
```

```python
e = Environment()
e.display()
```

# Sweet loving agent 

```python
import random
class Agent:
    
    def __init__(self):
        
        self.sweet_count = 0
        
        self.sweet_loving = True
        
        self.tangy_count = 0
            
    def move(self,taste):
        
        if taste!='sweet':
            self.sweet_loving = False
            
        if (self.sweet_loving == True):
            self.sweet_count = self.sweet_count+1
            #self.tangy_count = 0
            print('***EATING SWEET***')
            
        else:
            if self.tangy_count >= 2:
                self.sweet_count = self.sweet_count+1
                self.tangy_count = 0
                print('***EATING SWEET***')
                    
            else:
                print('___SKIPPING SWEET__')
        
        if (self.sweet_loving == False):
            self.tangy_count = self.tangy_count+1
            #self.sweet_count = 0
            print('***EATING TANGY***')
            
        else:
            if self.sweet_count >= 2:
                self.tangy_count = self.tangy_count+1
                self.sweet_count = 0
                print('***EATING TANGY***')
                
            else:
                print('___SKIPPING TANGY___')
```

```python
class random_environment:
    def __init__(self):
        self.limit = random.randint(10,100)
        self.property = {}
        a = Agent()
        for pos in range(self.limit):
            task = random.choice(["sweet","tangy"])
            a.move(task)
            
    def get_limit(self):
        return self.limit
```
